```python
# Re-making dataset
def fuc():

    #graph_lact = pd.read_excel('lact.xlsx', engine='openpyxl')
    graph_lact = pd.read_csv('dsc.csv',index_col=0)
    
    graph_iloc = graph_lact[0:12]
    print(graph_iloc.columns)
    print('      ')
    
    return graph_iloc

def lact_110(graph_iloc):
    df_110 = graph_iloc[['time','mn_110','mw_110','DPI_110']]
    df_110 
    return df_110

def lact_120(graph_iloc):
    df_120 = graph_iloc[['time','mn_120','mw_120','DPI_120']]
    df_120 
    return df_120

def lact_130(graph_iloc):
    df_130 = graph_iloc[['time','mn_130','mw_130','DPI_130']]
    df_130 
    return df_130

def simple_regression_no_test(lact_escolhido):
    
    import numpy as np
    import pandas as pd
    from pandas import Series,DataFrame

    import matplotlib.pyplot as plt
    import seaborn as sns

    import sklearn
    from sklearn.linear_model import LinearRegression

    sns.set_style('whitegrid')
    %matplotlib inline
    
    #df.iloc[:, 2]
    #df[df.columns[2]
    
    data_time = lact_escolhido.iloc[:, 0]
    data_mn = lact_escolhido.iloc[:, 1]
    data_mw = lact_escolhido.iloc[:, 2]
    data_dpi = lact_escolhido.iloc[:, 3]
    print('  ')
    
    #################################################################
    # creating footage/figure and axis with subplots()
    fig,ax = plt.subplots()
    
    # Making plot - Criando o plot
    ax.plot(data_time, data_mn, color="red", marker="o")
    
    # setting x-axis label
    ax.set_xlabel("Time",fontsize=14)
    # setting y-axis label
    ax.set_ylabel("Mn",color="red",fontsize=14)
    
    
    # twin object for two different y-axis on the sample plot
    ax2=ax.twinx()
    
    # make a plot with different y-axis using second axis object
    ax2.plot(data_time, data_dpi,color="blue",marker="o")
    ax2.set_ylabel("DPI",color="blue",fontsize=14)
    plt.title(label='Mn Vs DPI 110')
    plt.show()
    # save the plot as a file
    #fig.savefig('two_different_y_axis_for_single_python_plot_with_twinx.jpg',format='jpeg',dpi=100,bbox_inches='tight')

    ########################################################################
    
    return plt.show()
```

```python
x = fuc()
y = lact_110(x)
#lact_120(x)
#lact_130(x)
simple_regression_no_test(y)
```

```python
#x = fuc()
#y = lact_110(x)
#lact_120(x)
#lact_130(x)
#regression_no_test(y)

def regression_no_test(lact_escolhido):
    
    import numpy as np
    import pandas as pd
    from pandas import Series,DataFrame

    import matplotlib.pyplot as plt
    import seaborn as sns

    import sklearn
    from sklearn.linear_model import LinearRegression

    sns.set_style('whitegrid')
    %matplotlib inline
    
    #df.iloc[:, 2]
    #df[df.columns[2]
    
    data_time = lact_escolhido.iloc[:, 0]
    data_mn = lact_escolhido.iloc[:, 1]
    print(data_mn)  # Valores de X
    print('  ')
    
    data_mw = lact_escolhido.iloc[:, 2]
    print(data_mw)
    
    data_dpi = lact_escolhido.iloc[:, 3]
    print('  ')
    
    X = data_mn.values.reshape(-1, 1) # ([ , ],[ , ])
    Y = data_mw
    print('  ')
    
    
    # MODELO SEM!!!!!!!!!!! TEST VALORES
    ############################################################
    
    model = LinearRegression(fit_intercept=True) # fit_intercept=False
    model.fit(X,Y)
    print(model.fit(X,Y))
    print('  ')
    
    predictions = model.predict(X).dtype
    predictions = model.predict(X)
    print(predictions)
    print('  ')
    
    coefficient_y = model.intercept_
    print('coefficient_y (valor independente) ==>',coefficient_y)
    
    coefficient_x = model.coef_
    print('coefficient_x (valor dependente) ==>',coefficient_x)
    
    r2_score = model.score(X,Y)
    print(r2_score*100,'%')
    print('  ')
    
    
    # REGRESSAO COM OS VALORES DE Y E Y_PREDICTED
    #################################################################
    # creating footage/figure and axis with subplots()
    plt.style.use('default')
    plt.style.use('ggplot')
    
    fig,ax = plt.subplots()
    
    # Making plot - Criando o plot
    ax.plot(Y, predictions, color="red", marker="o")

    # setting x-axis label
    ax.set_xlabel("Y",fontsize=14)
    # setting y-axis label
    ax.set_ylabel("predictions",color="red",fontsize=14)
    
    # make a plot with different y-axis using second axis object
    plt.title(label='Y Vs predictions')
    plt.show()
    # save the plot as a file
    #fig.savefig('two_different_y_axis_for_single_python_plot_with_twinx.jpg',format='jpeg',dpi=100,bbox_inches='tight')
    ########################################################################
        
    
    
    
    
    # preprocessing required by scikit-learn functions
    # FAZENDO REGRESSAO LINEAR COM VALORES DE LINSPACE POR MEIO DE PREDICTIONS
    ################################################################# 
    x_valores_de_teste_para_regressao = np.linspace(200, 400, 30)            # 200 data points between 0 ~ 40
    x_pred = x_valores_de_teste_para_regressao.reshape(-1, 1)  # preprocessing required by scikit-learn functions
    
    predictions = model.predict(x_pred).dtype
    predictions = model.predict(x_pred)
    print(predictions)
    print('  ')
    
    
    plt.style.use('default')
    plt.style.use('ggplot')
    fig, ax = plt.subplots(figsize=(7, 3.5))
    ax.plot(x_pred, predictions, color='k', label='Regression model')
    ax.scatter(X, Y, edgecolor='k', facecolor='grey', alpha=0.7, label='Sample data')
    fig.tight_layout()
    

    return 
```

```python
x = fuc()
y = lact_110(x)
#lact_120(x)
#lact_130(x)
regression_no_test(y)
```

```python
#x = fuc()
#y = lact_110(x)
#lact_120(x)
#lact_130(x)
#regression_no_test(y)

def regression_COM_test(lact_escolhido):
    
    import numpy as np
    import pandas as pd
    from pandas import Series,DataFrame

    import matplotlib.pyplot as plt
    import seaborn as sns

    import sklearn
    from sklearn.linear_model import LinearRegression

    sns.set_style('whitegrid')
    %matplotlib inline
    
    #df.iloc[:, 2]
    #df[df.columns[2]
    
    data_time = lact_escolhido.iloc[:, 0]
    data_mn = lact_escolhido.iloc[:, 1]
    print(data_mn)  # Valores de X
    print('  ')
    
    data_mw = lact_escolhido.iloc[:, 2]
    print(data_mw)
    
    data_dpi = lact_escolhido.iloc[:, 3]
    print('  ')
    
    X = data_mn.values.reshape(-1, 1) # ([ , ],[ , ])
    Y = data_mw
    print('  ')
    
    # MODELO COM!!!!!!!!!!! TEST VALORES
    ############################################################
    from sklearn.model_selection import train_test_split
    from sklearn.linear_model import LinearRegression
    from sklearn.model_selection import train_test_split
    
    X_train, X_test, y_train, y_test = train_test_split(X ,Y, test_size=0.5, random_state=101)
   
    # MONTANDO MODELO COM X_TRAIN E Y_TRAIN VALORES
    model = LinearRegression(fit_intercept=True) # fit_intercept=False
    model.fit(X_train,y_train)
    print(model.fit(X_train,y_train))
    print('  ')
    
    # Modelo treinado com os valores testes - Model trained with test values
    Y_test_predict = model.predict(X_test).dtype
    Y_test_predict = model.predict(X_test)
    print(Y_test_predict)
    print('  ')
    
    coefficient_y = model.intercept_
    print('coefficient_y (valor independente) ==>',coefficient_y)
    
    coefficient_x = model.coef_
    print('coefficient_x (valor dependente) ==>',coefficient_x)
    
    r2_score = model.score(X_test,y_test)
    print(r2_score*100,'%')
    print('  ')
    
    print(Y_test_predict.shape)
    print(y_test.shape)
    
    # REGRESSAO COM OS VALORES DE Y_TEST E Y__TEST_PREDICTED
    #################################################################
    # creating footage/figure and axis with subplots()
    plt.style.use('default')
    plt.style.use('ggplot')
    
    fig,ax = plt.subplots()
    
    # Making plot - Criando o plot
    ax.plot(y_test, Y_test_predict, color="red", marker="o")

    # setting x-axis label
    ax.set_xlabel("y_test",fontsize=14)
    # setting y-axis label
    ax.set_ylabel("Y_test_predict",color="red",fontsize=14)
    
    # make a plot with different y-axis using second axis object
    plt.title(label='y_test Vs Y_test_predict')
    plt.show()
    # save the plot as a file
    #fig.savefig('two_different_y_axis_for_single_python_plot_with_twinx.jpg',format='jpeg',dpi=100,bbox_inches='tight')
    ########################################################################
        
    # Construcao da linha de regressao
    # preprocessing required by scikit-learn functions
    # FAZENDO REGRESSAO LINEAR COM VALORES DE LINSPACE POR MEIO DE PREDICTIONS
    ################################################################# 
    x_valores_de_teste_para_regressao = np.linspace(200, 400, 30)            # 200 data points between 0 ~ 40
    x_pred = x_valores_de_teste_para_regressao.reshape(-1, 1)  # preprocessing required by scikit-learn functions
    
    predictions = model.predict(x_pred).dtype
    predictions = model.predict(x_pred)
    print(predictions)
    print('  ')
    
    
    plt.style.use('default')
    plt.style.use('ggplot')
    fig, ax = plt.subplots(figsize=(7, 3.5))
    ax.plot(x_pred, predictions, color='k', label='Regression model')
    
    ax.scatter(X, Y, edgecolor='k', facecolor='grey', alpha=0.7, label='Sample data')
    
    # setting x-axis label
    ax.set_xlabel("Mn",fontsize=14)
    # setting y-axis label
    ax.set_ylabel("Mw",color="red",fontsize=14)
    
    plt.title(label='Mn Vs Mw')
    fig.tight_layout()
    

    return 
```
